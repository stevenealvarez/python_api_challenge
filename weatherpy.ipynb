```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np 
import requests
import time
import json
from scipy.stats import linregress

#api key
from api_key import weather_api_key

from citipy import citipy

output_data_file = "output_data/cities.csv"

#have to determine a range for list of cities based on range for lat and lng
lat_range = (-90,90)
lng_range = (-180,180)
```

```python
#create a list to hold values for cities and latitude and longitude 
lat_lng = []
cities= []


#need random cities based on the lat and lng 
lat = np.random.uniform(low=-90.000, high=90.00, size=1500)
lng = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lng = zip(lat,lng)

#have it go through the list created for random lat_lng
for lat_lng in lat_lng:
    city= citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    #so cities are not duplicated
    if city not in cities:
        cities.append(city)
#how many cities are in list.... print count needs to be over 500+      
len(cities)
```

```python
print("Starting Data Retrieval")
print("--------------------------")

url = "http://api.openweathermap.org/data/2.5/weather?"
units = "imperial"
query_url = f"{url}appid={weather_api_key}&units={units}&q="

responses = []
#loop through each city in list cities.... change 200 to 500 !!!
i = 0 
for city in cities: 
    i = i + 1 
    response = requests.get(query_url + city).json()
    if response['cod']== 200:
        print(f"Processing Record {i} | {city}")
        responses.append(response)
    else: 
        print("City not found. Skip.....")
print("Data Retrieval Complete")
```

```python
#need group of list to hold data response to convert raw data into data frame

city_name= []
country = []
date = []
latitude = []
longitude = []
max_temp = []
humidity = []
cloudiness = []
wind_speed = []

#go through each url/city to pull information required to create df
for resp in responses: 
    city_name.append(resp['name'])
    country.append(resp['sys']['country'])
    date.append(resp['dt'])
    humidity.append(resp['main']['humidity'])
    latitude.append(resp['coord']['lat'])
    longitude.append(resp['coord']['lon'])
    max_temp.append(resp['main']['temp_max'])
    cloudiness.append(resp['clouds']['all'])
    wind_speed.append(resp['wind']['speed'])
    
cities_dict = {
    'City': city_name,
    'Country': country,
    'Date': date,
    'Humidity': humidity,
    'Lat': latitude,
    'Lng': longitude,
    'Max Temp': max_temp,
    'Cloudiness': cloudiness,
    'Wind Speed': wind_speed
}

cities_data_df = pd.DataFrame(cities_dict)

cities_data_df.to_csv(output_data_file, index=False)

cities_data_df.head(10)
```

```python
#Linear Regressions...ex matplotlib data 5.3


#Northern Hemisphere Temp (F) v Lat
northern_hemisphere_df = cities_data_df.loc[cities_data_df['Lat'] > 0, :]
southern_hemisphere_df = cities_data_df.loc[cities_data_df['Lat'] < 0, :]

x_values = northern_hemisphere_df['Lat']
y_values = northern_hemisphere_df['Max Temp']
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(6,10),fontsize=15,color="red")
plt.xlabel('Latitude')
plt.ylabel('Max Temp (F)')
plt.title('Northern Max Temp (F) v Latitude Regression')
plt.show()

plt.savefig('output_data/N_TempVsLat.png')
```

#The linear regression demonstrates how the temperature drops as you move north of the equator.
